```r
library(tidyverse)
library(ggplot2)
library(gganimate)
library(gifski)
library(dplyr)

data <- read.csv("/kaggle/input/rent-price-in-barcelona-2014-2022/Barcelona_rent_price.csv")

head(data)
summary(data)
```

<div style="font-family:NewCenturySchoolbook;font-size:20px;">Now let's create a new dataframe with those values.</div>

```r
year <- c(2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022)
rent_per_month <- c(y_2014_month,y_2015_month, y_2016_month, y_2017_month, y_2018_month, y_2019_month, y_2020_month, y_2021_month, y_2022_month)
rent_sq_meter <- c(y_2014_meter,y_2015_meter, y_2016_meter, y_2017_meter, y_2018_meter, y_2019_meter, y_2020_meter, y_2021_meter, y_2022_meter)

rent_per_year <- data.frame(year,rent_per_month,rent_sq_meter)
rent_per_year
```

# <h1 style="font-family:NewCenturySchoolbook;font-size:20px;margin-left:35%;"> Dinamic view on rent prices </h1>

 <div style="font-family:NewCenturySchoolbook;font-size:20px;">And now I'd like to animate my plot and see how the prices have been moving dinamically and do some additional improvements to out plot</div>

```r
rent_price <- prices_per_district_long %>% mutate(district = gsub("_", " ", district)) %>% ggplot() + geom_point(aes(year, value, color=district), size=4, alpha=0.3) +
  ggtitle("Rent prices in different districts of \nBarcelona") +
  ylab("Price per month per m2 in Euro") +
  expand_limits(y=7.5) +
  labs(color="District", subtitle='Year: {frame_time}') +
  geom_segment(aes(x=2014, y=7.5, xend=2022, yend=7.5), linetype=2, color="gray") +
  geom_segment(aes(x=2014, y=15.0, xend=2022, yend=15.0), linetype=2, color="gray") +
  geom_segment(aes(x=2014, y=12.5, xend=2022, yend=12.5), linetype=2, color="gray") +
  geom_segment(aes(x=2014, y=10.0, xend=2022, yend=10.0), linetype=2, color="gray") +
  theme(plot.background = element_rect(fill="#FBF8F1"),
        panel.background = element_rect(fill="#FBF8F1"),
        legend.background = element_rect(fill="#FBF8F1"),
        legend.key = element_rect(fill="#FBF8F1"),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        plot.title = element_text(family="NewCenturySchoolbook", size=22, margin=margin(10,10,30,-10)),
        axis.title.x = element_blank(),
        axis.title.y = element_text(family="NewCenturySchoolbook", size=15, hjust=0.49),
        axis.ticks.x=element_blank(),
        axis.text.x=element_blank(),
        axis.text.y = element_text(family="NewCenturySchoolbook", size=10, margin=margin(0,10,0,10)),
        plot.subtitle = element_text(family="NewCenturySchoolbook", size=12),
        plot.margin = margin(0,1,0,1, "cm")) +
        transition_time(year)
        
anim_save("rent_price.gif", rent_price)
```
